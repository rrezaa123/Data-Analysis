```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'E:\Analysis Data\file\african_econ_crises.csv')

df.head(10)
```

```python
# cleaning data

missing_values = df.isnull().sum()

stats_descriptive = df.describe()

duplicated_rows = df[df.duplicated()]

data_types = df.dtypes

missing_values, stats_descriptive, duplicated_rows, data_types
```

```python
# Filter data Egypt
df_mesir = df[df['country'] == 'Egypt']
```

```python
# Plot bar for 'banking_crisis'
plt.figure(figsize=(8, 6))
df_mesir['banking_crisis'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Banking Crises in Egypt')
plt.xlabel('Banking Crisis Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Plot time series for 'inflation_annual_cpi'
plt.figure(figsize=(12, 6))
plt.plot(df_mesir['year'], df_mesir['inflation_annual_cpi'], marker='o', linestyle='-', color='b', label='Inflation (Annual CPI)')
plt.title('Annual Inflation in Egypt')
plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot scatter plot for the relationship between 'gdp_weighted_default' and 'exch_usd'
plt.figure(figsize=(8, 6))
plt.scatter(df_mesir['gdp_weighted_default'], df_mesir['exch_usd'], color='green', alpha=0.5)
plt.title('Relationship between GDP Weighted Default and Exchange Rate in Egypt')
plt.xlabel('GDP Weighted Default')
plt.ylabel('Exchange Rate (to USD)')
plt.grid(True)
plt.show()
```

```python
# Plot boxplots for multiple numeric columns
plt.figure(figsize=(10, 6))
df_mesir[['exch_usd', 'inflation_annual_cpi']].boxplot()
plt.title('Boxplot of Exchange Rate and Annual Inflation in Egypt')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
